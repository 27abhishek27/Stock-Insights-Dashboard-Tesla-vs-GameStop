```python
import yfinance as yf

# Step 1: Download Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="5y")  # Fetch data for the last 5 years

# Step 2: Reset the index
tesla_data.reset_index(inplace=True)

# Step 3: Save the DataFrame to a CSV file
tesla_data.to_csv('tesla_stock_data.csv', index=False)

# Step 4: Display the first 5 rows
print(tesla_data.head())
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
import pandas as pd

# Path to your chromedriver
driver_path = 'path_to_your_chromedriver'  # Update this path

# Step 1: Use the Service class to specify the driver path
service = Service(driver_path)

# Step 2: Open the browser and visit the page
driver = webdriver.Chrome(service=service)
url = 'https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue'
driver.get(url)

# Wait for the page to load completely (you may adjust the time if necessary)
time.sleep(5)

# Step 3: Extract the table
table = driver.find_element(By.XPATH, "//table")

# Step 4: Extract data from the table
rows = table.find_elements(By.TAG_NAME, "tr")
data = []

for row in rows[1:]:  # Skip the header row
    columns = row.find_elements(By.TAG_NAME, "td")
    if len(columns) >= 2:
        year = columns[0].text
        revenue = columns[1].text
        data.append([year, revenue])

# Step 5: Create a DataFrame
tesla_revenue = pd.DataFrame(data, columns=['Year', 'Revenue'])

# Step 6: Show the last 5 rows
print(tesla_revenue.tail())

# Close the browser
driver.quit()
```

```python
import yfinance as yf
import pandas as pd

# Step 1: Download GameStop's financial data (including revenue) from Yahoo Finance
gme = yf.Ticker("GME")
financials = gme.financials

# Step 2: Filter the revenue data (usually available under the 'Total Revenue' row)
revenue_data = financials.loc['Total Revenue']

# Step 3: Convert the data into a DataFrame for easier manipulation
revenue_df = pd.DataFrame(revenue_data).transpose()

# Step 4: Display the last 5 rows
print(revenue_df.tail())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have the GameStop stock data in a CSV file or a DataFrame
# For example, load data from a CSV file:
# df = pd.read_csv("GameStop_Stock_Data.csv")

# Example: Let's create a sample DataFrame with Date and Close columns for simplicity.
# In practice, this data will come from a CSV or API.

data = {
    'Date': ['2021-01-01', '2021-01-02', '2021-01-03', '2021-01-04'],
    'Close': [20.0, 50.0, 200.0, 300.0]
}

df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])

# Plotting function (make_graph)
def make_graph(dataframe):
    plt.figure(figsize=(10, 6))
    plt.plot(dataframe['Date'], dataframe['Close'], marker='o', color='b', label='GameStop Stock Price')
    plt.title('GameStop Stock Price Over Time')
    plt.xlabel('Date')
    plt.ylabel('Stock Price (USD)')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()

    # Show the plot
    plt.show()

# Call the function to plot the data
make_graph(df)
```
